```
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib 
```

```
data_dau_all = pd.read_csv('https://raw.githubusercontent.com/bakoyashi/python-dataAnalysis/master/dau.csv',
                           parse_dates=True, index_col=0)
data_dau = data_dau_all[['DAU', 'Android', 'iOS']]
data_dau = data_dau_all.resample('B').ffill()
data_dau.head()
```

```
data_dau.DAU.plot()
data_dau.DAU.rolling(28).mean().plot()
plt.title('DAUおよび28日移動平均DAU');
```

```
data_dau.rolling(28).mean().plot(logy=True)
plt.rcParams['ytick.direction'] = 'out'
plt.title('DAUおよびデバイス別DAU');
```

```
dau_px = data_dau.DAU['2018':'2019']
ma60 = dau_px.rolling(30, min_periods=20).mean()
ewma60 = dau_px.ewm(span=30).mean()
ma60.plot(style='b--', label='単純移動平均')
ewma60.plot(style='b-', label='指数移動平均')
plt.legend()
plt.title('60日DAU(単純移動平均と指数移動平均の比較)');
```
