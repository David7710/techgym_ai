```
import pandas as pd
import codecs

pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
with codecs.open("/content/drive/My Drive/Colab Notebooks/Chapter15/datasets/movielens/users.dat", "r", "utf-8", "ignore") as file:
     users = pd.read_csv(file, sep='::',
                       header=None, names=unames, engine='python')
```

```
users[:5]
```

```
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
with codecs.open("/content/drive/My Drive/Colab Notebooks/Chapter15/datasets/movielens/ratings.dat", "r", "utf-8", "ignore") as file:
     ratings = pd.read_csv(file, sep='::',
                       header=None, names=rnames, engine='python')
```

```
ratings[:5]
```

```
mnames = ['movie_id', 'title', 'genres']
with codecs.open("/content/drive/My Drive/Colab Notebooks/Chapter15/datasets/movielens/movies.dat", "r", "utf-8", "ignore") as file:
    movies = pd.read_csv(file, sep='::',
                       header=None, names=mnames, engine='python')
```

```
movies[:5]
```

```
# 3つのデータをマージする
data = pd.merge(pd.merge(ratings, users), movies)
data[:5]
```

```
data.iloc[0]
```

```
#性別・年齢ごとの平均評価を出す
mean_ratings = data.pivot_table('rating', index='title',
                                columns='gender', aggfunc='mean')
mean_ratings[:5]
```

```
#各映画タイトルでレビュー数を集計
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```
#レビュー数が250以上の映画を抽出
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles[:10]
```

```
#インデックスに存在する行のみを選択する
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings[:10]
```

```
#女性評価の高いものの上位10件
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

```
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```
#女性の評価が高く、男性の評価が低かった映画
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:10]
```

```
#男性の評価が高く、女性の評価が低かった映画
sorted_by_diff[::-1][:10]
```

```
#性別は丸めて、評価が大きく割れた映画を調べる。

# 映画タイトルごとの評価の標準偏差
rating_std_by_title = data.groupby('title')['rating'].std()
# 評価数250件以上の映画のみを抽出
rating_std_by_title = rating_std_by_title.loc[active_titles]
# 得られたオブジェクトを降順でソート
rating_std_by_title.sort_values(ascending=False)[:10]
```
