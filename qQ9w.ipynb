```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib 
```

```
bills = pd.read_csv('https://raw.githubusercontent.com/bakoyashi/python-dataAnalysis/master/bills.csv')
counts = pd.crosstab(bills['day'],bills['locale'])
```

```
counts.plot.bar()
plt.title('曜日と言語別の課金回数')
```

```
bills['bills_pct'] = bills['support'] / (bills['total_bill'] - bills['support'])
sns.barplot(x='bills_pct', y='day', data=bills, orient='h')
plt.title('課金額のうち「サポート」の割合　[曜日別]')
```

```
sns.barplot(x='bills_pct', y='day', hue='locale', data=bills, orient='h')
plt.title('課金額のうち「サポート」の割合　[曜日&言語]')
```

```
bills['bills_pct'].plot.hist(bins=50)
plt.title('課金額のうち「サポート」の割合のヒストグラム')
```

```
bills['bills_pct'].plot.density()
plt.title('課金額のうち「サポート」の割合の密度プロット')
```
